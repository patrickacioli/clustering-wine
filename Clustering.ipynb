### KMeans

Uma das mÃ©tricas comumente usadas para comparar os resultados em diferentes valores de K Ã© a distÃ¢ncia mÃ©dia entre os pontos de dados e seu centrÃ³ide de cluster. Como o aumento do nÃºmero de clusters sempre reduzirÃ¡ a distÃ¢ncia atÃ© os pontos de dados, o aumento de K sempre diminuirÃ¡ essa mÃ©trica, atÃ© o extremo de atingir zero quando K for igual ao nÃºmero de pontos de dados. Assim, essa mÃ©trica nÃ£o pode ser usada como o Ãºnico destino. Em vez disso, a distÃ¢ncia mÃ©dia para o centrÃ³ide em funÃ§Ã£o de K Ã© traÃ§ada e o "ponto de cotovelo" ou "elbow point", onde a taxa de diminuiÃ§Ã£o muda acentuadamente, pode ser usado para determinar aproximadamente K.

```python
inertias = []
for k in range(1,20):
    kmeans_model = KMeans(n_clusters=k).fit(x_scaled)
    inertias.append(kmeans_model.inertia_)
```

```python
plt.plot(range(1,20), inertias, "--", color="black")
plt.plot(range(1,20), inertias, "o", color="red")
plt.xlabel("Valor de K")
plt.ylabel("InÃ©rcia")
plt.title("Valores de K")
plt.annotate("Ponto de cotovelo, K = %s" % str(3), xy=(3, inertias[2]),xytext=(5, 1500), 
             arrowprops=dict(shrink=0.05))
```

```python
kmeans_model = KMeans(n_clusters=3).fit(x_scaled)
```

```python
plt.figure()
sns.scatterplot(x="proline", y="OD280", hue=kmeans_model.labels_, data=df)
plt.title("KMeans Clustering")
```

### Hierarchical AgglomerativeClustering

O critÃ©rio de ligaÃ§Ã£o (linkage) determina qual distÃ¢ncia usar entre conjuntos de observaÃ§Ã£o. O algoritmo mesclarÃ¡ os pares de cluster que minimizam esse critÃ©rio.


- ward minimiza a variaÃ§Ã£o dos clusters que estÃ£o sendo mesclados.
- mÃ©dia usa a mÃ©dia das distÃ¢ncias de cada observaÃ§Ã£o dos dois conjuntos.
- complete ou usa as distÃ¢ncias mÃ¡ximas entre todas as observaÃ§Ãµes dos dois conjuntos.
- single usa o mÃ­nimo das distÃ¢ncias entre todas as observaÃ§Ãµes dos dois conjuntos.

```python
for link in ["ward", "complete", "average", "single"]:
    plt.figure()
    agg_cluster = AgglomerativeClustering(linkage=link).fit(x_scaled)    
    sns.scatterplot(x="proline", y="OD280", hue=agg_cluster.labels_, data=df)
    plt.title("AgglomerativeClustering Clustering -> %s" % link)
    plt.text(1200,1.5, "n_clusters=%s" % agg_cluster.n_clusters_)
```

### DBSCAN

A distÃ¢ncia mÃ¡xima entre duas amostras para uma ser considerada na vizinhanÃ§a da outra. Este nÃ£o Ã© um limite mÃ¡ximo nas distÃ¢ncias de pontos dentro de um cluster. Este Ã© o parÃ¢metro DBSCAN mais importante para escolher apropriadamente para o seu conjunto de dados e funÃ§Ã£o de distÃ¢ncia.

Neste notebook, variou-se o parÃ¢metro eps de 2 atÃ© 3 e verificou-se que os valores de 2.2 e 2.4 conseguiram se performar proÃ³ximo a classificaÃ§Ã£o original do dataset (3 classes)

```python
for i in np.arange(2, 3, 0.2):
    plt.figure()
    dbscan_cluster = DBSCAN(eps=i).fit(x_scaled)    
    sns.scatterplot(x="proline", y="OD280", hue=dbscan_cluster.labels_, data=df)
    plt.title("DBSCAN Clustering -> eps=%.02f" % i)
    plt.text(1200,1.5, "n_clusters=%s" % len(np.unique(dbscan_cluster.labels_)))
```
