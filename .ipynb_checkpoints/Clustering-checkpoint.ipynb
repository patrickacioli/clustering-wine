### KMeans

Uma das métricas comumente usadas para comparar os resultados em diferentes valores de K é a distância média entre os pontos de dados e seu centróide de cluster. Como o aumento do número de clusters sempre reduzirá a distância até os pontos de dados, o aumento de K sempre diminuirá essa métrica, até o extremo de atingir zero quando K for igual ao número de pontos de dados. Assim, essa métrica não pode ser usada como o único destino. Em vez disso, a distância média para o centróide em função de K é traçada e o "ponto de cotovelo" ou "elbow point", onde a taxa de diminuição muda acentuadamente, pode ser usado para determinar aproximadamente K.

```python
inertias = []
for k in range(1,20):
    kmeans_model = KMeans(n_clusters=k).fit(x_scaled)
    inertias.append(kmeans_model.inertia_)
```

```python
plt.plot(range(1,20), inertias, "--", color="black")
plt.plot(range(1,20), inertias, "o", color="red")
plt.xlabel("Valor de K")
plt.ylabel("Inércia")
plt.title("Valores de K")
plt.annotate("Ponto de cotovelo, K = %s" % str(3), xy=(3, inertias[2]),xytext=(5, 1500), 
             arrowprops=dict(shrink=0.05))
```

```python
kmeans_model = KMeans(n_clusters=3).fit(x_scaled)
```

```python
plt.figure()
sns.scatterplot(x="proline", y="OD280", hue=kmeans_model.labels_, data=df)
plt.title("KMeans Clustering")
```

### Hierarchical AgglomerativeClustering

O critério de ligação (linkage) determina qual distância usar entre conjuntos de observação. O algoritmo mesclará os pares de cluster que minimizam esse critério.


- ward minimiza a variação dos clusters que estão sendo mesclados.
- média usa a média das distâncias de cada observação dos dois conjuntos.
- complete ou usa as distâncias máximas entre todas as observações dos dois conjuntos.
- single usa o mínimo das distâncias entre todas as observações dos dois conjuntos.

```python
for link in ["ward", "complete", "average", "single"]:
    plt.figure()
    agg_cluster = AgglomerativeClustering(linkage=link).fit(x_scaled)    
    sns.scatterplot(x="proline", y="OD280", hue=agg_cluster.labels_, data=df)
    plt.title("AgglomerativeClustering Clustering -> %s" % link)
    plt.text(1200,1.5, "n_clusters=%s" % agg_cluster.n_clusters_)
```

### DBSCAN

A distância máxima entre duas amostras para uma ser considerada na vizinhança da outra. Este não é um limite máximo nas distâncias de pontos dentro de um cluster. Este é o parâmetro DBSCAN mais importante para escolher apropriadamente para o seu conjunto de dados e função de distância.

Neste notebook, variou-se o parâmetro eps de 2 até 3 e verificou-se que os valores de 2.2 e 2.4 conseguiram se performar proóximo a classificação original do dataset (3 classes)

```python
for i in np.arange(2, 3, 0.2):
    plt.figure()
    dbscan_cluster = DBSCAN(eps=i).fit(x_scaled)    
    sns.scatterplot(x="proline", y="OD280", hue=dbscan_cluster.labels_, data=df)
    plt.title("DBSCAN Clustering -> eps=%.02f" % i)
    plt.text(1200,1.5, "n_clusters=%s" % len(np.unique(dbscan_cluster.labels_)))
```

### Guassian Mixture

O objeto GaussianMixture implementa o algoritmo de expectativa de maximização (EM) para ajustar modelos de mistura de Gauss. Ele também pode desenhar elipsóides de confiança para modelos multivariados e calcular o Bayesian Information Criterion para avaliar o número de clusters nos dados. 

A GaussianMixture vem com diferentes opções para restringir a covariância das classes de diferença estimadas: covariância full, tied, diag ou spherical.

- full: cada componente tem sua própria matriz de covariância geral
- tiel: todos os componentes compartilham a mesma matriz de covariância geral
- diag: cada componente tem sua própria matriz de covariância diagonal
- spherical: cada componente tem sua própria variância única
